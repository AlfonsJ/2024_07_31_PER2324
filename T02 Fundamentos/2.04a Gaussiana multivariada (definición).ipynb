# Distribución Gaussiana multivariada

## Función de densidad de probabilidad

$$p(\boldsymbol{y}\mid\boldsymbol{\mu},\boldsymbol{\Sigma})=%
\frac{1}{(2\pi)^{D/2}\lvert\boldsymbol{\Sigma}\rvert^{1/2}}%
\exp\left[-\frac{1}{2}(\boldsymbol{y}-\boldsymbol{\mu})^t\boldsymbol{\Sigma}^{-1}(\boldsymbol{y}-\boldsymbol{\mu})\right]$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
def plot_norm2d(m, S):
    x, y = np.mgrid[-5:5:.1, -5:5:.1]
    z = multivariate_normal(m, S).pdf(np.dstack((x, y)))
    fig = plt.figure(figsize=(16, 8))
    ax = fig.add_subplot(121, projection='3d')
    ax.plot_surface(x, y, z, color='white', edgecolor="black")
    ax = fig.add_subplot(122, aspect='equal')
    ax.grid()
    ax.axhline(0, color='black')
    ax.axvline(0, color='black')
    ax.contour(x, y, z)
    plt.show()
```

```python
m, S = [0, 0], [[1, 0], [0, 1]]
plot_norm2d(m, S)
```

```python
m, S = [0, 0], [[1, 0], [0, 3]]
plot_norm2d(m, S)
```

```python
m, S = [0, 0], [[2., 1.8], [1.8, 2.]]
plot_norm2d(m, S)
```
